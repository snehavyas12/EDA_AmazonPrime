### **Step 2: Handling Missing Values.**

The first step in handling missing data is to identify its presence and magnitude. This can be done using the command "df.isnull().sum()", which provides a count of missing entries for each column. Once we know where the missing values are concentrated, the next step is to remove any unnecessary or irrelevant columns that contain excessive null values or add little analytical value. This can be done using "df.drop(columns = [...])". After dropping, it is good practice to recheck the dataset with "df.isnull().sum()" to confirm which variables still contain nulls.

For the remaining missing values, the treatment depends on whether the column is categorical or numerical. In the case of categorical variables, missing entries are often filled with the most frequent category (mode), as this preserves the dominant pattern in the data. The mode can be identified using "df['column_name'].mode()" and then applied with "fillna()".

For numerical variables, the choice of imputation method—whether to use the mean or the median, depends on the distribution of the data. If the data is normally distributed, the mean is preferred since it accurately represents the central tendency. However, if the data is skewed, the mean may be influenced by extreme values (outliers), and therefore the median becomes a better choice. To make this decision, we first visualize the distribution using tools like histograms or boxplots, and then quantitatively measure skewness using the "skew()" function.

A skew value close to zero suggests a normal distribution where the mean is suitable. A skew value less than 0.5 still indicates a fairly normal spread, so the mean can be used. If the skewness lies between 0.5 and 1.0, the distribution is moderately skewed, and it is often safer to use the median. Finally, if the skewness is greater than 1 or less than -1, the distribution is highly skewed, and the median should definitely be chosen to avoid distortion from outliers.

By following this systematic process—identifying nulls, dropping irrelevant columns, and then applying thoughtful imputation strategies—we ensure that the dataset remains consistent, reliable, and ready for accurate analysis without compromising its underlying structure.


```python
 #out of 15 columns, 8 columns have null values in them
titles_df.isnull().sum() #gives total number of null values in each column
```

```python
titles_df.drop(['description', 'age_certification', 'seasons', 'imdb_id', 'imdb_votes'], axis = 1, inplace = True)
```

```python
titles_df.isnull().sum() #after dropping unnecessary columns, we have 4 column with null values
```

```python
#as all the columns have numerical datatypes. we need to fill them with their descriptive statistics
#to determine the type of average we should use to fill the null values, we look at the distribution of the variable.
title_cols = ['imdb_score', 'tmdb_popularity', 'tmdb_score']
for col in title_cols:
  sns.boxplot(x = titles_df[col])
  plt.title(f"Boxplot of {col}")
  plt.show()
```

```python
#to capture the spread of the distribution more accurately, we calculate the skewness of the data in each column
print(titles_df[title_cols].skew())
```

```python
#any column with > 0.5 is highly skewed and should use median as its average, else we use mean.
title_cols = ['imdb_score', 'tmdb_popularity', 'tmdb_score']
for col in title_cols:
  if abs(titles_df[col].skew()) > 0.5:
    titles_df[[col]] = titles_df.loc[:, [col]].fillna(titles_df.loc[:, [col]].median())
  else:
    titles_df[[col]] = titles_df.loc[:, [col]].fillna(titles_df.loc[:, [col]].mean())
```

##### **Q3. What is the average IMDB score for TV shows? Does length of runtime vary the IMDb ratings?**

```python
average_imdb = titles_df.groupby('type')['imdb_score'].mean().reset_index()
average_imdb_shows = round(average_imdb[average_imdb['type'] == 'SHOW'], 2)
average_imdb_shows
```

```python
# Step 1: Create 10–minute runtime bins
shows_df = titles_df[titles_df['type'] == 'SHOW'].copy()

bin_edges = range(10, shows_df['runtime'].max() + 10, 10)  # from 0 to max runtime, step of 10
shows_df['runtime_bin'] = pd.cut(shows_df['runtime'], bins=bin_edges)

# Step 2: Group by those bins and calculate average IMDb score
avg_imdb_for_runtime = shows_df.groupby('runtime_bin')['imdb_score'].mean().reset_index()

# Step 3: Handle nulls (if any remain)
avg_imdb_for_runtime['imdb_score'] = avg_imdb_for_runtime['imdb_score'].fillna(avg_imdb_for_runtime['imdb_score'].mean())

# Step 4: Round off the results
avg_imdb_for_runtime = avg_imdb_for_runtime.round(2)
print(avg_imdb_for_runtime)
```

```python
#Convert bins to midpoints for plotting
fig, ax = plt.subplots(figsize = (10, 4))
avg_imdb_for_runtime['bin_mid'] = avg_imdb_for_runtime['runtime_bin'].apply(lambda x: x.mid)

plt.plot(avg_imdb_for_runtime['bin_mid'], avg_imdb_for_runtime['imdb_score'], marker='o')
plt.xlabel("Runtime (minutes)")
plt.ylabel("Average IMDb Score")
plt.title("Average IMDb Score by Runtime Bin")
plt.show()
```

Why Line Chart?
Line Chart is the most appropraite chart for continuous variable. We have runtime bins, at interval 10. Especially with many runtime bins, a line plot avoids the clutter of bars and highlights the overall trajectory rather than just counts.

Insights:
As assumed, the runtime does not have extreme impact on the IMDB ratings. Some episodes are lengthy, others are not, it is predominantly the quality of content that impacts the IMDB ratings. This is because the range of imdb score is 6.0 - 7.75.
